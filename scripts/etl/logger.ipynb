```python
# ======================================================================
#  Unified Logger Setup (Notebook version of scripts/etl/logger.py)
# ======================================================================

import logging

def setup_logger(name: str):
    """
    Create and configure a logger with a unique name to track ETL steps.

    Works inside Jupyter notebooks—prints messages directly
    in the notebook output cells.
    """

    logger = logging.getLogger(name)
    logger.setLevel(logging.INFO)

    # avoid adding handlers repeatedly on re‑imports
    if not logger.handlers:
        handler = logging.StreamHandler()
        formatter = logging.Formatter(
            "[%(asctime)s] %(levelname)s - %(name)s - %(message)s"
        )
        handler.setFormatter(formatter)
        logger.addHandler(handler)

    return logger


# ----------------------------------------------------------------------
# Example usage inside the notebook
# ----------------------------------------------------------------------

logger = setup_logger("notebook_logger")
logger.info("Logger initialized successfully ✅ (Notebook Example)")
```
